# MÃ©todo del Gradiente

```python
import numpy as np
import matplotlib.pyplot as plt

# Definamos la funciÃ³n
def f(x, y):
    return -(4*x**2) - (57*y**2) + 4*x*y - 8*x + 10*y - 2

# Definamos el gradiente de la funciÃ³n
def grad_f(x, y):
    return np.array([-8*x + 4*y - 8, -114*y + 4*x + 10])

# Definamos los parÃ¡metros del mÃ©todo
point = np.array([2, -5]) # Punto semilla
precision = 0.00001 # Nivel de precisiÃ³n
step_size = 0.1 # Longitud de paso

# Ahora creamos la figura
fig, ax = plt.subplots(figsize=(8, 6))

# Creamos la malla de puntos para graficar la funciÃ³n
x_range = np.linspace(-10, 10, 100)
y_range = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x_range, y_range)
Z = f(X, Y)

# Graficamos la funciÃ³n
ax.contour(X, Y, Z, levels=20, cmap='coolwarm')

# Iteramos hasta alcanzar el nivel de precisiÃ³n
while np.linalg.norm(grad_f(point[0], point[1])) > precision:
    # Calcular el gradiente en el punto actual
    grad = grad_f(point[0], point[1])
    # Calcular la direcciÃ³n del mÃ¡ximo aumento de la funciÃ³n
    direction = -grad / np.linalg.norm(grad)
    # Actualizar el punto avanzando en la direcciÃ³n opuesta
    point = point + step_size * direction
    # Graficar el punto actual
    ax.plot(point[0], point[1], 'bo')
    plt.pause(0.1) # Pausar para mostrar la iteraciÃ³n en la grÃ¡fica

# Mostramos el punto crÃ­tico en la grÃ¡fica
ax.plot(point[0], point[1], 'ro')

# Mostramos la grÃ¡fica
plt.show()
```
